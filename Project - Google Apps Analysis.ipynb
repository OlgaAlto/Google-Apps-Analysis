```python
# Delite dublicats

apps = apps.drop_duplicates(keep='last')
apps
```

```python
#checking for total null values

apps.isnull().sum() 
```

```python
# check a type in columns

apps.dtypes
```

```python
apps[apps['Price'] != '0'].head() 
```

```python
# Delite "$" in strings

apps.Price = apps.Price.str.replace("$", "").astype(float)
```

```python
apps.Price[apps.Price >0]
```

```python
# Rename column

apps = apps.rename(columns={'Price': 'Price in $'})
apps[apps['Price in $'] >0].head(3)
```

```python
apps.Installs.head(20)
```

```python
# Cleaning "+" and "," to ""

apps.Installs = apps.Installs.str.replace("+", "").str.replace(",", "").astype(float)
```

```python
apps.Installs
```

```python
# Replace "Varies with device" and "M" to ""

apps.Size = apps.Size.str.replace("Varies with device", "")
apps.Size = apps.Size.str.replace("M", "")
```

```python
#How many elements end with "k"

apps.Size.str.endswith("k").value_counts()
```

```python
# elements index that end with "k"

apps[apps.Size.str.endswith("k")].index
```

```python
 apps.loc[apps.Size.str.contains('k'), 'Size'] = apps.loc[apps.Size.str.contains('k'), 'Size'].str.replace('k', '').astype(float) *0.001
```

```python
# Randon checking that size "KB" converted to "M"

apps.Size[58]
```

```python
# Plot by Category

apps['Category'].value_counts().plot(kind='bar');
plt.title('Plot by Category')

# RESULT: Apps in "Family" category is the most popular.
```

```python
# Plot "Price"

hist_by_Type = apps.Type.hist()
plt.title('Comparison of paid and free applications')

# RESULT: The biggest part of apps is free.
```

```python
# Plot 'Content Rating by group'

apps['Content Rating'].value_counts().plot(kind='bar');
plt.title('Content Rating by group')

# RESULT: the most popular apps in group "Everyone".
```

```python
# Subset by Category

db = apps.groupby(["Category"]).agg({'Installs':[min,max], 'Reviews':max})
db
```

```python
# Regression analysis: The higher the rating, the more reviews

sns.regplot(x="Rating", y="Reviews", data=apps)

# RESULT: Rating has strong relationship with Reviews.
```

```python
# Plot 'Reviews' by 'Category'

db = apps.groupby(["Category"]).agg({'Reviews': 'max'})
fig, ax = plt.subplots(figsize=(16, 6))
feature_2 = db.plot(kind='line', y='Reviews',ax=ax, secondary_y=True)

# RESULT: The most amount of reviews in categories "Productivity" and "Comics".
```

```python
# Do free apps have more installs than paid apps on average? 

apps.groupby(['Type']).mean()

# RESULT: YES, free apps have more installs than paid apps on average.
```

```python
# Which categories appear the most among top 100 most expensive apps?

df1 = pd.DataFrame()
df1 = apps.sort_values(by='Price in $', ascending=False).head(100)

df1.groupby("Category").count()

# RESULT: Top 100 most expensive apps in "Famyly" and "Medical" categories.
```

```python
# The top 10 most expensive apps
apps.nlargest(columns='Price in $', n=10)


# RESULT: The top 10 most expensive apps in everage cost 400$.
```

```python
# top 10 most common words that appear in apps name

from collections import Counter
from itertools import chain

n = 10
individual_components = chain.from_iterable(apps['App'].str.split(', '))
counter = Counter(individual_components)
print(counter.most_common(n))
```

```python
# Correlation Between Variables 

data = apps[['Price in $','Rating', "Installs"]]
correlation = data.corr(method='pearson')
correlation

# RESULT: The obtained correlation coefficients are very small and mean that there are no connections between the studied objects.
```

```python
# Pie chart by 'Content Rating'

cc = apps['Content Rating'].value_counts().plot(kind='pie', figsize=(10,5), autopct="%.0f%%", pctdistance = 0.75, textprops={'fontsize': 14})

cc.set_title('Content Rating', fontsize=16)
cc.set_ylabel('')
circle = plt.Circle((0,0), 0.6, color='white')
cc.add_artist(circle)
plt.axis('equal')
plt.tight_layout()

# RESULT: The most popular group of  apps is for "Everyone".
```
